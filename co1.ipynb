```python
def multiply_arrays(arr1, arr2):
    if len(arr1) != len(arr2):
        raise ValueError("Arrays must be of the same size")
    return [a * b for a, b in zip(arr1, arr2)]


arr1 = [1, 2, 3]
arr2 = [4, 5, 6]
print(multiply_arrays(arr1, arr2))
```

```python
def calculator():
    print("Select operation: +, -, *, /")
    op = input("Enter operation: ")
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    
    if op == '+':
        print("Result:", num1 + num2)
    elif op == '-':
        print("Result:", num1 - num2)
    elif op == '*':
        print("Result:", num1 * num2)
    elif op == '/':
        if num2 == 0:
            print("Cannot divide by zero")
        else:
            print("Result:", num1 / num2)
    else:
        print("Invalid operation")

calculator()
```

```python
def arrays_equal(arr1, arr2):
    if len(arr1) != len(arr2):
        return False
    return all(a == b for a, b in zip(arr1, arr2))


print(arrays_equal([1, 2, 3], [1, 2, 3])) 
print(arrays_equal([1, 2, 3], [1, 2, 4]))  
```

```python
import numpy as np


A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])


C = A + B
print("Addition:\n", C)


D = np.dot(A, B)
print("Multiplication:\n", D)


print("Transpose of A:\n", A.T)


U, S, VT = np.linalg.svd(A)
print("U:\n", U)
print("Singular values:\n", S)
print("VT:\n", VT)
```

```python
import numpy as np
import matplotlib.pyplot as plt

men_scores = [22, 30, 35, 35, 26]
women_scores = [25, 32, 30, 35, 29]

groups = ['G1', 'G2', 'G3', 'G4', 'G5']
x = np.arange(len(groups))
width = 0.35

fig, ax = plt.subplots()
bar1 = ax.bar(x - width/2, men_scores, width, label='Men')
bar2 = ax.bar(x + width/2, women_scores, width, label='Women')

ax.set_xticks(x)
ax.set_xticklabels(groups)
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```
